```python
# calculate the probability mass function of 2 coin tosses
```

```python
import random
def coin_toss(n):
  outcomes = ["Heads", "Tails"]
  result = []
  for x in range(n):
    outcome = random.choice(outcomes)
    result.append(outcome)
  return result

result = coin_toss(1000)
print(result)

total_toss = len(result)
head_toss = result.count("Heads")
tail_toss = result.count("Tails")

print("Total toss: ", total_toss)
print("Head toss: ", head_toss)
print("Tail toss: ", tail_toss)


def plot_pmf(result):
  import matplotlib.pyplot as plt
  import numpy as np
  unique, counts = np.unique(result, return_counts=True)
  plt.bar(unique, counts/len(result))

  plt.title("Probability Mass Function")
  plt.xlabel("Outcome")
  plt.ylabel("Probability")
  plt.show()

plot_pmf(result)

```

```python
# probability density function of a normal distribution

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = 1/(np.sqrt(2*np.pi))*np.exp(-x**2/2)

plt.plot(x, y)
```
