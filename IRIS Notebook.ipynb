```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
df = pd.read_csv(url)
df

```

```python
df.describe()
```

```python
df.sort_values(by='sepal_length', ascending=False)
```

```python
groupby_species = df.groupby('species')
groupby_species.mean()
```

```python
pivot = df.pivot_table(index='species', values='petal_length', aggfunc =['mean', 'min', 'max'])
pivot
```

```python
import matplotlib.pyplot as plt
species_colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
df['color'] = df['species'].map(species_colors)
plt.scatter(df['sepal_length'],df['petal_length'],c = df['color'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df, hue='species')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.show()
```

```python
violin = sns.violinplot(x='species', y='petal_length', data=df)
plt.show()
```

```python
sns.histplot(df['sepal_length'], kde=True)
plt.show()
#
```

```python
#histogram by species
sns.histplot(data=df, x='sepal_length', hue='species', kde=True)
plt.show()
#
```

```python
variables = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for variable in variables:
  g=sns.FacetGrid(df)
  g.map(sns.histplot, variable, kde=True, color = "skyblue")
  plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
data
```

```python
#Data correlations

data = data.select_dtypes(include=['number'])
correlation_matrix = data.corr()
correlation_matrix


```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
